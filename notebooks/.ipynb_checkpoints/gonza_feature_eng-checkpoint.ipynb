```python
from numpy import loadtxt
#from xgboost import XGBClassifier
#from sklearn.model_selection import train_test_split
#from sklearn.metrics import accuracy_score
import pandas as pd
import seaborn as sns
```

```python
events = pd.read_csv('../dataset/events_up_to_01062018.csv', low_memory = False)
test = pd.read_csv('../dataset/labels_training_set.csv')
```

```python
events.head()
```

```python
#CREO UN DATA FRAME CON LAS PERSONAS PARA IR JOINEANDO MIS FEATURES AHI
persons = events['person'].to_frame().drop_duplicates()
persons.info()
```

```python
#ELIMINO COLUMNA MODEL YA QUE TENEMOS EL SKU QUE ES UN VALOR NUMERICO (MEJOR PARA LOS ALGORITMOS DE ML) REPRESENTANDO
#EL MODELO
events = events.drop(columns='model')

#TRANSFROMO LA FECHA A TIMESTAMP
events['timestamp'] = pd.to_datetime(events['timestamp'])
```

```python
#AGREGO UNA COLUMNA DE CANTIDAD
events['cantidad'] = 1
```

```python
#HACIENDO ESTE GROUPBY NOS QUEDAMOS CON LOS SKUS QUE VIO UNA PERSONA Y EN QUE FECHAg
grouped_sku = events.groupby(['person','sku']).agg({'cantidad':'sum'})
grouped_sku.sort_values(by='cantidad', ascending=False, inplace=True)
#grouped_sku.drop_duplicates(inplace=True, , keep='first')
grouped_sku = grouped_sku.reset_index().drop_duplicates(keep='first', subset='person')
grouped_sku.drop(columns='cantidad', inplace=True)
grouped_sku.columns = ['person', 'most_viewed_sku']
grouped_sku.info()
persons = pd.merge(persons, grouped_sku, how='left', on='person')
persons.info()
```

```python
conversions = events.loc[events['event'] == 'conversion']
#checkouts = events.loc[events['event'] == 'checkout']
#viewed_products = events.loc[events['event'] == 'viewed product']
#sns.countplot(data=conversions, x='sku', order=conversions.sku.value_counts().head(5).index)
```

```python
#DATAFRAME DE USUARIOS QUE REALIZARON UNA CONVERSION
grouped_conversions = conversions.groupby(['person']).agg({'cantidad':'count'})
grouped_conversions.reset_index(inplace=True)
person_converted = grouped_conversions.drop(columns='cantidad')
person_converted['has_converted_before'] = 1
persons = pd.merge(person_converted, persons, how='right', on='person')
```

```python
#HACIENDO ESTE GROUPBY NOS QUEDAMOS CON LOS SKUS QUE VIO UNA PERSONA Y EN QUE FECHAg
grouped_color = events.groupby(['person','color']).agg({'cantidad':'sum'})
grouped_color.sort_values(by='cantidad', ascending=False, inplace=True)
#grouped_sku.drop_duplicates(inplace=True, , keep='first')
grouped_color = grouped_color.reset_index().drop_duplicates(keep='first', subset='person')
grouped_color.drop(columns='cantidad', inplace=True)
grouped_color.columns = ['person', 'most_viewed_color']
grouped_color.info()
persons = pd.merge(persons, grouped_color, how='left', on='person')
persons.info()
```

```python
grouped_region = events.groupby(['person','region']).agg({'cantidad':'sum'})
grouped_region.sort_values(by='cantidad', ascending=False, inplace=True)
grouped_region = grouped_region.reset_index().drop_duplicates(keep='first', subset='person')
grouped_region.drop(columns='cantidad', inplace=True)
grouped_region.columns = ['person', 'region_brasil']
persons = pd.merge(persons, grouped_region, how='left', on= 'person')
```

```python
persons
```

```python
persons.fillna(value=0, inplace=True)
persons_dummies = pd.get_dummies(persons, columns=['most_viewed_color', 'region_brasil'])
persons_filtered = persons_dummies[['person',
                                    'has_converted_before','most_viewed_sku','most_viewed_color_Preto',
                                   'most_viewed_color_Branco','most_viewed_color_Dourado', 'region_brasil_Sao Paulo',
                                   'region_brasil_Unknown', 'region_brasil_Minas Gerais']]
```

```python
persons_filtered.to_csv('../features/features_gonza.csv')
```
