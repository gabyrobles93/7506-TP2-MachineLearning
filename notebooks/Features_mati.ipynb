Vamos a calificar cuales usuarios son mas propensos a tener eventos de compras, estos eventos van a ser los eventos tipo; conversion, checkout.

Lo podemos dividir en compradores:
    1)Frecuentes - Promedio mayor a 0.6
    2)Esporadicos - Promedio mayor a 0.2  menor a 0.6
    3)Casi nunca - Promedio menor a 0.2
    3)Nunca - Promedio 0

```python
person_checkout_conversion = df.loc[(df.event == "conversion") | (df.event == "checkout") ]
person_checkout_conversion["CantidadConv"] = (person_conversion.event == "conversion")
person_checkout_conversion["CantidadChekout"] = (person_conversion.event == "checkout")
person_checkout_conversion = person_checkout_conversion.groupby("person").agg({"CantidadConv":"sum", "CantidadChekout":"sum"}).reset_index()
person_checkout_conversion
```

# Promedio de conversiones respecto a los checkout

```python
person_checkout_conversion["promedio"] = (person_checkout_conversion["CantidadConv"])/(person_checkout_conversion["CantidadChekout"])
person_checkout_conversion
```

```python
person_frecuencia = person_checkout_conversion.loc[:,["person","promedio"]]
person_frecuencia["promedio"] = person_frecuencia.promedio.astype('float64')
person_frecuencia
```

```python
person_frecuencia["Frecuente"] = person_frecuencia.promedio >= 0.6
person_frecuencia["Esporadicos"] = (person_frecuencia.promedio > 0.2) & (person_frecuencia.promedio < 0.6)
person_frecuencia["Casi nunca"] = (person_frecuencia.promedio <= 0.2)  & (person_frecuencia.promedio > 0)
person_frecuencia["Nunca"] = (person_frecuencia.promedio == 0)
person_frecuencia
```

```python
otro = person_frecuencia.loc[:, ["person","Frecuente","Esporadicos","Casi nunca","Nunca"]]
```

```python
otro
```

```python
otro["Frecuencia"] = "nada"
otro.loc[(otro["Esporadicos"] == True), "Frecuencia"] = "Esporadico"
otro.loc[(otro["Frecuente"] == True), "Frecuencia"] = "Frecuente"
otro.loc[(otro["Casi nunca"] == True), "Frecuencia"] = "Casi nunca"
otro.loc[(otro["Nunca"] == True), "Frecuencia"] = "Nunca"

otro.loc[:, ["person","Frecuencia"]]
```
