## Vamos a calificar cuales usuarios son mas propensos a tener eventos de compras, estos eventos van a ser los eventos tipo; conversion, checkout.
Lo podemos dividir en compradores:

    1) Frecuentes - Promedio mayor a 0.6
    2) Esporadicos - Promedio mayor a 0.2  menor a 0.6
    3) Casi nunca - Promedio menor a 0.2
    4) Nunca - Promedio 0

```python
person_checkout_conversion = df
person_checkout_conversion["CantidadConv"] = (person_conversion.event == "conversion")
person_checkout_conversion["CantidadChekout"] = (person_conversion.event == "checkout")
person_checkout_conversion = person_checkout_conversion.groupby("person").agg({"CantidadConv":"sum", "CantidadChekout":"sum"}).reset_index()
person_checkout_conversion
```

# Promedio de conversiones respecto a los checkout

```python
person_checkout_conversion["promedio"] = (person_checkout_conversion["CantidadConv"])/(person_checkout_conversion["CantidadChekout"])
person_checkout_conversion.fillna(value=0, inplace=True)
person_checkout_conversion
```

```python
person_checkout_conversion.sort_values(by="CantidadConv", ascending=False)
```

```python
person_frecuencia = person_checkout_conversion.loc[:,["person","promedio"]]
person_frecuencia["promedio"] = person_frecuencia.promedio.astype('float64')
person_frecuencia
```

```python
person_frecuencia["Frecuente"] = person_frecuencia.promedio >= 0.6
person_frecuencia["Esporadicos"] = (person_frecuencia.promedio > 0.2) & (person_frecuencia.promedio < 0.6)
person_frecuencia["Casi_nunca"] = (person_frecuencia.promedio <= 0.2)  & (person_frecuencia.promedio > 0)
person_frecuencia["Nunca"] = (person_frecuencia.promedio == 0)
person_frecuencia
```

```python
otro = person_frecuencia.loc[:, ["person","Frecuente","Esporadicos","Casi_nunca","Nunca"]]
```

```python
otro
```

```python
otro["Frecuencia"] = "nada"
otro.loc[(otro["Esporadicos"] == True), "Frecuencia"] = "Esporadico"
otro.loc[(otro["Frecuente"] == True), "Frecuencia"] = "Frecuente"
otro.loc[(otro["Casi_nunca"] == True), "Frecuencia"] = "Casi_nunca"
otro.loc[(otro["Nunca"] == True), "Frecuencia"] = "Nunca"

otro = otro.loc[:, ["person","Frecuencia"]]

otro
```

```python
features_mati_encoded = pd.get_dummies(otro, columns=['Frecuencia'])
```

```python
features_mati_encoded.to_csv('features_frecuencia.csv', index=False)
```

## En busqueda de alguna relación con las campañas adheridas

```python
campania_usuarios = df.loc[df.event == "ad campaign hit"]
```

```python
campania_usuarios = campania_usuarios.loc[:,["person", "campaign_source"]]
campania_usuarios["Cantidad_De_Campaña"] = 1
campania_usuarios.groupby(["person","campaign_source"]).agg({"Cantidad_De_Campaña":"count"}).reset_index()
```

```python
cant_de_campanias = campania_usuarios.groupby(["person"]).agg({"Cantidad_De_Campaña":"count"}).reset_index()

mergeo = pd.merge(cant_de_campanias,otro, on="person",how="inner")
```

```python
mergeo.columns = ["persona","Cant_Campañas","Frecuencia"]
mergeo
```

```python
mergeo.groupby("Frecuencia").agg({"persona":"count","Cant_Campañas":"sum"}).reset_index()
```
